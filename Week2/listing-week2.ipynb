## Missing Value Analysis
- count each columns and sum all the missing values 
- calculate the missing values ratios
- if proportion > p then treat as high missing col
- flag columns with >90% missing values

```python
# highmissing threshold
p_high = 0.5
p_very_high = 0.9

listing_missing_summary = pd.DataFrame({
    'missing_count': listing.isna().sum(),
    'missing_ratio': listing.isna().mean()
}).sort_values('missing_ratio', ascending=False)

listing_high_missing_summary = listing_missing_summary[listing_missing_summary['missing_ratio'] > p_high]

listing_high_missing_summary['very_high_missing_flag'] = (listing_high_missing_summary['missing_ratio'] > p_very_high).astype(int)

listing_high_missing_summary
```

## Numerical Distribution Review
Analysis the distribution of key numeric fields include
ClosePrice, ListPrice, OriginalListPrice, LivingArea,
LotSizeAcres, BedroomsTotal, BathroomsTotalInteger, DaysOnMarket, and YearBuilt. 

For each field, generate histograms, boxplots, and percentile summaries, and identify extreme outliers for later handling.

```python
# create a list of numeric fields to analyze for outliers
listing_numeric_fields = ['ClosePrice','ListPrice','OriginalListPrice','LivingArea','LotSizeAcres','BedroomsTotal','BathroomsTotalInteger','DaysOnMarket','YearBuilt']

listing_outlier_summary = []

for col in listing_numeric_fields:
    # define which fields to log-transform for better visualization
    log_fields = ['ClosePrice','ListPrice','OriginalListPrice','LivingArea','LotSizeAcres','DaysOnMarket']
    raw_fields = ['BedroomsTotal','BathroomsTotalInteger','YearBuilt']

    s = pd.to_numeric(listing[col], errors='coerce').dropna()

    print(f"\n===== {col} =====")

    percentile_summary = s.describe(percentiles=[0.01,0.05,0.25,0.5,0.75,0.95,0.99])
    print(percentile_summary.apply(lambda x: f'{x:,.2f}'))

    # calculate IQR and outlier thresholds
    q1 = s.quantile(0.25)
    q3 = s.quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr

    # define the extreme outlier thresholds
    extreme_lower = q1 - 3 * iqr
    extreme_upper = q3 + 3 * iqr

    outliers = s[(s < lower_bound) | (s > upper_bound)]
    extreme_outliers = s[(s < extreme_lower) | (s > extreme_upper)]

    listing_outlier_summary.append({
        'field': col,
        'count': len(s),
        'missing_count': listing[col].isna().sum(),
        'missing_ratio': listing[col].isna().mean(),
        'min': s.min(),
        'p1': s.quantile(0.01),
        'p5': s.quantile(0.05),
        'p25': q1,
        'p50': s.quantile(0.50),
        'p75': q3,
        'p95': s.quantile(0.95),
        'p99': s.quantile(0.99),
        'max': s.max(),
        'iqr': iqr,
        'outlier_count': len(outliers),
        'outlier_ratio': len(outliers) / len(s),
        'extreme_outlier_count': len(extreme_outliers),
        'extreme_outlier_ratio': len(extreme_outliers) / len(s),
        'lower_bound': lower_bound,
        'upper_bound': upper_bound,
        'extreme_lower_bound': extreme_lower,
        'extreme_upper_bound': extreme_upper
    })
    
    plt.figure(figsize=(10, 2.5))
    
    if col in log_fields:
        sns.boxplot(x=np.log1p(s))
        plt.xlabel(f'log(1 + {col})')
        plt.title(f'Boxplot of log-transformed {col}')

        plt.figure(figsize=(10, 4))
        sns.histplot(x=np.log1p(s), bins=30, kde=True)
        plt.title(f'Histogram of log-transformed {col}')
        plt.xlabel(f'log(1 + {col})')
        plt.ylabel('Frequency')

    else:
        sns.boxplot(x=s)
        plt.xlabel(col)
        plt.title(f'Boxplot of {col}')

        plt.figure(figsize=(10, 4))
        sns.histplot(s, bins=30, kde=True)
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)
        plt.ylabel('Frequency')

    
    plt.show()

listing_outlier_summary = pd.DataFrame(listing_outlier_summary)
listing_outlier_summary
```
